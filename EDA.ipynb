Checking for missing values

```python
# Check for missing values
missing_values = df.isnull().sum()

# Calculate percentage of missing values
missing_percentage = (missing_values / len(df)) * 100

# Create a DataFrame to display the missing values and their percentage
missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})

# Display the missing data
print(missing_data[missing_data['Missing Values'] > 0])
```

```python
import numpy as np
import pandas as pd

# Calculate the Z-scores of all numerical columns
z_scores = np.abs(df.select_dtypes(include=[np.number]).apply(lambda x: (x - x.mean()) / x.std()))

# Define a threshold for identifying outliers
threshold = 3

# Find where Z-scores are greater than the threshold
outliers = z_scores > threshold

# Count the number of outliers in each column
outliers_count = outliers.sum().sort_values(ascending=False)

# Print the counts of outliers in each column
print("Number of outliers in each column:")
print(outliers_count)
```

 'Fiesta' these outliers could represent peak event days that are important for our analysis

Data Cleaning 

```python
# Convert the 'Dia' column to datetime
df['Dia'] = pd.to_datetime(df['Dia'], format='%Y-%m-%d')

# Convert columns with commas to float
df['Latitud'] = df['Latitud'].str.replace(',', '.').astype(float)
df['Longitud'] = df['Longitud'].str.replace(',', '.').astype(float)
df['PVP'] = df['PVP'].str.replace(',', '.').astype(float)
df['VentasN'] = df['VentasN'].str.replace(',', '.').astype(float)

# Display the first few rows after conversion
print(df.head())
```

### Visualizations


Target Variable

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plotting the time series of sales
plt.figure(figsize=(15, 6))
plt.plot(df['Dia'], df['VentasN'], label='Daily Sales')
plt.title('Time Series of Sales')
plt.xlabel('Dia')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Calculate the z-score to identify outliers
from scipy.stats import zscore

df['z_scores'] = zscore(df['VentasN'])
df = df[df['z_scores'].abs() <= 3]  # Adjust the threshold as necessary

# Plotting again
plt.figure(figsize=(15, 6))
plt.plot(df['Dia'], df['VentasN'], label='Daily Sales')
plt.title('Time Series of Sales (Outliers Removed)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Smoothing the data using a rolling mean
df['smoothed_sales'] = df['VentasN'].rolling(window=7, center=True).mean()  # 7-day window

plt.figure(figsize=(15, 6))
plt.plot(df['Dia'], df['smoothed_sales'], label='Smoothed Sales', color='orange')
plt.title('Smoothed Time Series of Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Aggregating sales by month
df['Month'] = df['Dia'].dt.to_period('M')
monthly_sales = df.groupby('Month')['VentasN'].sum()

plt.figure(figsize=(15, 6))
plt.plot(monthly_sales.index.astype(str), monthly_sales.values, label='Monthly Sales')
plt.title('Monthly Time Series of Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

Decomposition of Time Series

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(df['VentasN'], model='additive', period=30)  # based on monthly data
result.plot()
plt.show()
```

Rolling Statistics

```python
rolling_window = 30  # 30 days rolling window
df['rolling_mean'] = df['VentasN'].rolling(window=rolling_window).mean()
df['rolling_std'] = df['VentasN'].rolling(window=rolling_window).std()

plt.figure(figsize=(15, 6))
plt.plot(df['VentasN'], label='Original')
plt.plot(df['rolling_mean'], label='Rolling Mean')
plt.plot(df['rolling_std'], label='Rolling Std Deviation')
plt.title('Rolling Mean and Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.legend()
plt.show()
```

### Time Series Analysis

Check for Stationarity

Plot ACF and PACF

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Perform ADF test
result = adfuller(df['VentasN'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
for key, value in result[4].items():
    print(f'Critical Value ({key}): {value}')

# Plot ACF
plt.figure(figsize=(12, 6))
plot_acf(df['VentasN'], lags=30)
plt.show()

# Plot PACF
plt.figure(figsize=(12, 6))
plot_pacf(df['VentasN'], lags=30)
plt.show()
```
